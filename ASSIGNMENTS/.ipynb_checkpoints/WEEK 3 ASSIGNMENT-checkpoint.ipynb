```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("insurance2.csv")
df
```

**1. Remove the rows which is having NULL Values.**

```python
new_df = df.dropna()
print(new_df.to_string())
```

**2. Fill the Empty cells in the â€œchargesâ€ column with**
   - Mean
   - Median  
   - Mode

```python
x = df["charges"].mean()
df["charges"].fillna(x, inplace = True)
df
```

```python
x = df["charges"].median()
df["charges"].fillna(x, inplace = True)
df
```

```python
x = df["charges"].mode()
df["charges"].fillna(x, inplace = True)
df
```

**3. Replace the values which is more than 60 in the â€œageâ€ column with the value of 50.**

```python
for x in df.index:
    if df.loc[x, "age"] > 60:
        df.loc[x, "age"] = 50
df["age"]
```

**4. Find out the duplicates in the dataset and remove.**

```python
print(df.duplicated())
df.drop_duplicates(inplace = True)
```

**5. Find out the correlation between the data.**

```python
print(df.corr())
```

**6. Draw a scatter plot between â€œageâ€ and â€œbmiâ€**

```python
df.plot(kind = 'scatter', x = 'age', y = 'bmi')
plt.show()
```

**7. Draw a histogram plot on â€œageâ€ column**

```python
df["age"].plot(kind = 'hist')
```
